```python
def find_key(data, target_key):
    results = []
    if isinstance(data, dict):
        for key, value in data.items():
            if key == target_key:
                results.append(value)
            else:
                results.extend(find_key(value, target_key))
    elif isinstance(data, list):
        for item in data:
            results.extend(find_key(item, target_key))
    return results
```

```python
def find_key_path(data, target_key):
    """
    جستجو در دیکشنری تو در تو برای یک کلید مشخص و برگرداندن
    مسیر root-to-leaf به صورت دیکشنری نگهدارنده ساختار.
    """
    if isinstance(data, dict):
        for key, value in data.items():
            if key == target_key:
                return {key: value}
            elif isinstance(value, (dict, list)):
                found = find_key_path(value, target_key)
                if found:
                    return {key: found}  # مسیر تا کلید هدف
    elif isinstance(data, list):
        for item in data:
            found = find_key_path(item, target_key)
            if found:
                return found
    return None
```

```python
import requests
import json
import math


SUBGRAPH_ID = "5zvR82QoaXYFyDEKLZ9t6v9adgnptxYpKpSbxtgVENFV"

URL = f"https://gateway.thegraph.com/api/{GRAPH_API_KEY}/subgraphs/id/{SUBGRAPH_ID}"


# ----------------------------
# Functions
# ----------------------------

def run_query(query):
    r = requests.post(URL, json={"query": query})
    data = r.json()
    if "errors" in data:
        raise Exception(data["errors"])
    return data["data"]


def tick_to_price(tick, decimals0, decimals1):
    return 1.0001 ** tick * 10 ** (decimals0 - decimals1)


def sqrtPriceX96_to_price(sqrtPriceX96, decimals0, decimals1):
    p = (int(sqrtPriceX96) / (2**96))**2
    return p * 10 ** (decimals0 - decimals1)


# ----------------------------
# MAIN SCRIPT
# ----------------------------

def compute_lp_apr(
        pool_address,
        capital_usd,
        price_low,
        price_high
    ):

    # Query pool info
    query_pool = f"""
    {{
      pool(id: "{pool_address.lower()}") {{
        id
        token0 {{ symbol decimals }}
        token1 {{ symbol decimals }}
        feeTier
        liquidity
        sqrtPrice
        volumeUSD
        feesUSD
      }}
    }}
    """

    data = run_query(query_pool)
    pool = data["pool"]

    dec0 = int(pool["token0"]["decimals"])
    dec1 = int(pool["token1"]["decimals"])

    feeTier = int(pool["feeTier"])
    L_pool = int(pool["liquidity"])

    sqrtP = int(pool["sqrtPrice"])
    P_curr = sqrtPriceX96_to_price(sqrtP, dec0, dec1)

    print("\n--------------------------------")
    print("Pool Info Loaded")
    print("--------------------------------")
    print("Token0:", pool["token0"]["symbol"])
    print("Token1:", pool["token1"]["symbol"])
    print("Fee Tier:", feeTier / 10000, "%")
    print("Current price:", P_curr)
    print("--------------------------------")

    # ----------------------------
    # Compute how much liquidity your capital buys
    # ----------------------------
    fee = feeTier / 1e6
    volume24 = float(pool["volumeUSD"])
    fees24h = volume24 * fee

    # Your share of liquidity relative to TVL:
    # We estimate TVL using liquidity based method near current price:
    # Not 100% exact but standard approximation
    # TVL ≈ liquidity * (sqrt(P_high) - sqrt(P_low))
    sp_low = math.sqrt(price_low)
    sp_high = math.sqrt(price_high)
    delta = sp_high - sp_low

    TVL_est = L_pool * delta

    your_share = capital_usd / TVL_est

    # daily fees proportional to share
    your_daily_fees = your_share * fees24h

    # APR
    apr = (your_daily_fees * 365) / capital_usd * 100

    print("\n--------------------------------")
    print("Fee APR Estimation")
    print("--------------------------------")
    print(f"Estimated TVL in your range: ${TVL_est:,.2f}")
    print(f"Your share of liquidity: {your_share*100:.4f}%")
    print(f"Your daily fees: ${your_daily_fees:.2f}")
    print(f"APR: {apr:.2f}%")
    print("--------------------------------")

    return apr


# ----------------------------
# RUN EXAMPLE
# ----------------------------
if __name__ == "__main__":
    #pool = input("Pool address: ").strip().lower()
    capital = float(input("Capital USD: "))
    low = float(input("Lower price: "))
    high = float(input("Upper price: "))

    compute_lp_apr(pool_address, capital, low, high)
```

```python
import requests, json, math
from datetime import datetime

# تنظیمات
#API_KEY = "YOUR_API_KEY"   # اگر از gateway استفاده میکنی، در غیر این صورت از public subgraph URL استفاده کن
SUBGRAPH_ID = "5zvR82QoaXYFyDEKLZ9t6v9adgnptxYpKpSbxtgVENFV"
URL = f"https://gateway.thegraph.com/api/{GRAPH_API_KEY}/subgraphs/id/{SUBGRAPH_ID}"
# یا اگر میخوای با api.thegraph.com امتحان کنی (ممکنه deprecated باشه):
# URL = "https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v3"

def run_query(q):
    r = requests.post(URL, json={"query": q})
    r.raise_for_status()
    return r.json()

def analyze_pool(pool_address):
    pool = pool_address.lower()
    q = f"""
    {{
      pool(id: "{pool}") {{
        id
        totalValueLockedUSD
        volumeUSD
        feesUSD
        feeTier
      }}
      poolDayDatas(first: 2, orderBy: date, orderDirection: desc, where: {{ pool: "{pool}" }}) {{
        date
        tvlUSD
        volumeUSD
        feesUSD
      }}
    }}
    """
    data = run_query(q)
    if "errors" in data:
        print("Graph errors:", data["errors"])
        return
    dat = data["data"]
    if not dat["pool"]:
        print("Pool not found in subgraph.")
        return

    pool_info = dat["pool"]
    days = dat["poolDayDatas"]
    print(json.dumps(pool_info, indent=2))
    print("poolDayDatas (most recent first):")
    print(json.dumps(days, indent=2))

    # compute fee APR from last-day volume
    fee_tier = int(pool_info["feeTier"])
    fee_rate = fee_tier / 1_000_000.0   # e.g., 3000 -> 0.003
    # pick last day data if present
    if len(days) >= 1:
        day0 = days[0]
        volume_24h = float(day0["volumeUSD"])
        fees_24h = float(day0["feesUSD"]) if day0.get("feesUSD") else volume_24h * fee_rate
        tvl = float(day0.get("tvlUSD") or pool_info.get("totalValueLockedUSD") or 0)
    else:
        # fallback to pool-level aggregated numbers (less accurate)
        volume_24h = float(pool_info.get("volumeUSD") or 0)  # lifetime volume, not 24h — best effort
        fees_24h = volume_24h * fee_rate
        tvl = float(pool_info.get("totalValueLockedUSD") or 0)

    fee_apr = (fees_24h / tvl) * 365 * 100 if tvl > 0 else None

    # approximate total apr:
    # idea: approximate net change in TVL between today and yesterday (captures price moves + net inflows)
    approx_total_apr = None
    if len(days) >= 2:
        day1 = days[1]
        tvl_today = float(day0["tvlUSD"])
        tvl_yesterday = float(day1["tvlUSD"])
        # daily pnl for LP ~= (tvl_today - tvl_yesterday) + fees_24h
        daily_pnl = (tvl_today - tvl_yesterday) + fees_24h
        approx_total_apr = (daily_pnl / tvl_yesterday) * 365 * 100 if tvl_yesterday > 0 else None

    # Print results
    print("\n--- Results ---")
    print(f"Fee rate (decimal): {fee_rate}")
    print(f"Volume (last day): {volume_24h:,.2f} USD")
    print(f"Fees (last day): {fees_24h:,.2f} USD")
    print(f"TVL (used): {tvl:,.2f} USD")
    print(f"Fee APR (annualized from last day): {fee_apr:.2f}%")
    if approx_total_apr is not None:
        print(f"Approx Total APR (using TVL change + fees): {approx_total_apr:.2f}%")
        print(f"Difference (fee_apr - approx_total_apr): {fee_apr - approx_total_apr:.2f} pct points")
    else:
        print("Not enough day data to compute approx total APR (need 2 poolDayDatas).")

    return {
        "fee_apr": fee_apr,
        "approx_total_apr": approx_total_apr,
        "volume_24h": volume_24h,
        "fees_24h": fees_24h,
        "tvl": tvl,
        "days": days
    }

# usage example:
# res = analyze_pool("0x5ae13baaef0620fdae1d355495dc51a17adb4082")
# print(res)
```
